```python
df["Valeur fonciere"] = df["Valeur fonciere"].str.replace(",", ".")
df["Valeur fonciere"] = pd.to_numeric(df["Valeur fonciere"], errors="coerce")
```

```python
df = df[df["Valeur fonciere"].notna()]
df = df[df["Surface reelle bati"].notna()]
df = df[df["Surface reelle bati"] > 0]
df = df[df["Valeur fonciere"] > 10000]
```

```python
len(df)
```

```python
df["prix_m2"] = df["Valeur fonciere"] / df["Surface reelle bati"]
df = df[df["prix_m2"] < 20000]  # valeurs très élevées
df = df[df["prix_m2"] > 500]    # valeurs très basses
```

```python
len(df)
```

```python
df.isnull().sum()
```

```python
cols_to_drop = ["B/T/Q", "Prefixe de section", "1er lot", "No Volume", "Nature culture", "Nature culture speciale", "Identifiant de document","Reference document","Identifiant local","Surface terrain"]
df = df.drop(columns=cols_to_drop)
```

```python
df.isnull().sum()
```

```python
df["Type de voie"] = df["Type de voie"].fillna("Autre")
df["Code postal"] = df["Code postal"].fillna(0)  # ou '00000'
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=[col for col in df.columns if "Articles CGI" in col])
```

```python
lots_cols = [col for col in df.columns if "Surface Carrez" in col or "lot" in col]
df = df.drop(columns=lots_cols)
```
